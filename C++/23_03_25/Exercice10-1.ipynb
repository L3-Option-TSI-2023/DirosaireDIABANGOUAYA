# [C++] 10 - Bases de la programmation orientée objet - Exercice 1

1.   Ecrire une classe Point2d qui comporte deux variables membres de type double m_x et m_y initialisées à 0. Implémenter un constructeur et une fonction print() pour que le programme suivant affiche :

  Point2d(0, 0)

  Point2d(3, 4)

2.   Ajouter une fonction membre distanceA() qui prend un autre point en paramètre et calcule la distance euclidienne entre eux

2.   Ajouter une fonction non membre amie distanceEntre() qui calcule la distance euclidienne entre deux points passés en paramètres


```python
%%writefile main.cpp

#include <cmath>
#include <iostream>

// CLass du point 2 dimension
class Point2d 
{

  private:
    double m_x = {0.0}, m_y = {0.0};

  public:
    Point2d(double a = 0.0, double b = 0.0): m_x(a), m_y(b){}

  void print() const {
      std::cout << "Point2d(" << m_x << ", " << m_y << ")\n";
  }

  // Méthode pour le calil de la distance entre les points 
  double distanceA(const Point2d& anotherPoint) const {

    double distanceFromX = {0.0}, distanceFromY = {0.0}, totalDistance = {0.0};

    distanceFromX = m_x - anotherPoint.m_x;
    distanceFromY = m_y - anotherPoint.m_y;
    totalDistance = std::sqrt(distanceFromX*distanceFromX + distanceFromY*distanceFromY);

    return totalDistance;
  }

  friend double distanceEntre(const Point2d& p1, const Point2d& p2);
};

double distanceEntre(const Point2d& p1, const Point2d& p2) {
  
  double distanceFromX = {0.0},  distanceFromY = {0.0}, totalDistance = {0.0};
  distanceFromX = p1.m_x - p2.m_x;
  distanceFromY = p1.m_y - p2.m_y;

  totalDistance = std::sqrt(distanceFromX*distanceFromX + distanceFromY*distanceFromY);
  
  return totalDistance;
}

int main()
{
    // Question 1
    Point2d p1{};
    Point2d p2{ 3.0, 4.0 };
    p1.print();
    p2.print();
 
    // Question 2
    std::cout << "Distance entre deux points: " << p1.distanceA(p2) << '\n';
    
    // Question 3
    std::cout << "Distance entre deux points: " << distanceEntre(p1, p2) << '\n';

    return 0;
}
```
