1.   Ecrire deux classes Pomme et Banane qui dérivent de la même classe Fruit. Fruit devra avoir deux membres : un nom et une couleur.
Le programme suivant devrait compiler :
```c
int main()
{
	Pomme a{ "rouge" };
	Banane b{};

	std::cout << "Ma " << a.getNom() << " est " << a.getCouleur() << ".\n";
	std::cout << "Ma " << b.getNom() << " est " << b.getCouleur() << ".\n";

	return 0;
}
```
et produire le résultat :
```
Ma pomme est rouge.
Ma banane est jaune.
```

2.   Ajouter une nouvelle classe appelée GrannySmith qui dérive de Pomme. 
Le programme suivant devrait compiler :
```c
int main()
{
	Pomme a{ "rouge" };
	Banane b{};
	GrannySmith c;

	std::cout << "Ma " << a.getNom() << " est " << a.getCouleur() << ".\n";
	std::cout << "Ma " << b.getNom() << " est " << b.getCouleur() << ".\n";
	std::cout << "Ma " << c.getNom() << " est " << c.getCouleur() << ".\n";

	return 0;
}
```
et produire le résultat :
```
Ma pomme est rouge.
Ma banane est jaune.
Ma pomme granny smith est verte.
```


````python
%%writefile main.cpp

#include <iostream>
#include <string>

class Fruit {

	public:
			Fruit(std::string name, std::string color) : name_(name), color_(color) {}
			std::string getNom() { return name_; }
			std::string getCouleur() { return color_; }
	
	private:
			std::string name_;
			std::string color_;
};

class Pomme : public Fruit {
	public:
			Pomme(std::string color) : Fruit("pomme", color) {}
};

class Banane : public Fruit {
	public:
			Banane() : Fruit("banane", "jaune") {}
};

class GrannySmith : public Pomme {
	public:
		GrannySmith(std::string color) : Pomme("granny smith", color) {}
};

int main()
{
    Pomme a{ "rouge" };
    Banane b{};
		GrannySmith c;

    std::cout << "Ma " << a.getNom() << " est " << a.getCouleur() << ".\n";
    std::cout << "Ma " << b.getNom() << " est " << b.getCouleur() << ".\n";
		std::cout << "Ma " << c.getNom() << " est " << c.getCouleur() << ".\n";

    return 0;
}
````
