1.   Ecrire une classe template Paire1 qui permet de gérer des paires de valeurs de même type. Le programme suivant devrait fonctionner :
```c
int main()
{
	Paire1<int> p1_1 { 5, 8 };
	std::cout << "Paire: " << p1_1.first() << ' ' << p1_1.second() << '\n';

	const Paire1<double> p1_2 { 2.3, 4.5 };
	std::cout << "Paire: " << p1_2.first() << ' ' << p1_2.second() << '\n';

	return 0;
}
```
et afficher :
```
Paire: 5 8
Paire: 2.3 4.5
```

3.   Une paire chaine-valeur est un type particulier de paire où la première valeur est toujours de type chaine et la seconde de type quelconque. Ecrire une classe template PaireChaineValeur qui dérive de la classe partiellement spécialisée Paire. Le programme suivant devrait fonctionner :
```c
int main()
{
	PaireChaineValeur<int> pcv { "Hello", 5 };
	std::cout << "Paire: " << pcv.first() << ' ' << pcv.second() << '\n';

	return 0;
}
```
et afficher :
```
Paire: Hello 5
```

````python
%%writefile main.cpp
#include <iostream>

template<typename T>
class Paire1
{
private:
    T m_first;
    T m_second;

public:
    Paire1(const T& premier, const T& second)
        : m_first{ premier }, m_second{ second }
    {
    }

    const T& first() const { return m_first; }
    const T& second() const { return m_second; }
};

int main()
{
	Paire1<int> p1_1 { 5, 8 };
	std::cout << "Paire: " << p1_1.first() << ' ' << p1_1.second() << '\n';

	const Paire1<double> p1_2 { 2.3, 4.5 };
	std::cout << "Paire: " << p1_2.first() << ' ' << p1_2.second() << '\n';

	return 0;
}
````

````python
%%writefile main.cpp

#include <iostream>

template<typename T1, typename T2>
class Paire2
{
private:
    T1 m_valeur1;
    T2 m_valeur2;

public:
    Paire2(const T1& valeur1, const T2& valeur2)
        : m_valeur1{valeur1}, m_valeur2{valeur2}
    {
    }

    const T1& first() const { return m_valeur1; }
    const T2& second() const { return m_valeur2; }
};

int main()
{
    Paire2<int, double> p2_1 { 5, 6.7 };
    std::cout << "Paire: " << p2_1.first() << ' ' << p2_1.second() << '\n';

    const Paire2<double, int> p2_2 { 2.3, 4 };
    std::cout << "Paire: " << p2_2.first() << ' ' << p2_2.second() << '\n';

    return 0;
}
````

````python
%%writefile main.cpp

#include <iostream>
#include <string>

template <typename T>
class Paire {
private:
    T m_premiere;
    T m_deuxieme;
public:
    Paire(const T& premiere, const T& deuxieme)
        : m_premiere(premiere), m_deuxieme(deuxieme)
    {
    }

    const T& premiere() const { return m_premiere; }
    const T& deuxieme() const { return m_deuxieme; }
};

template <typename T>
class PaireChaineValeur : public Paire<std::string> {
public:
    PaireChaineValeur(const std::string& premiere, const T& deuxieme)
        : Paire<std::string>(premiere, std::to_string(deuxieme))
    {
    }

    const T& second() const { return std::stoi(Paire<std::string>::deuxieme()); }
};


int main()
{
PaireChaineValeur<int> pcv { "Hello", 5 };
std::cout << "Paire: " << pcv.premiere() << ' ' << pcv.deuxieme() << '\n';

return 0;
}
````
