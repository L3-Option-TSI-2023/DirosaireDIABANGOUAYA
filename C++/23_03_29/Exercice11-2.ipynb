# [C++] 11 - Surcharge d'opérateurs - Exercice 2

Reprendre le code de l'exercice [C++] 11-1 pour ajouter une surcharge des opérateurs ```<<``` et ```>>```. Le code suivant devrait compiler :

```c
int main()
{
	Fraction f1;
	std::cout << "Entrer la fraction 1: ";
	std::cin >> f1;

	Fraction f2;
	std::cout << "Entrer la fraction 2: ";
	std::cin >> f2;

	std::cout << f1 << " * " << f2 << " est " << f1 * f2 << '\n'; 

	return 0;
}
```
et produire l'affichage suivant :
```
Entrer la fraction 1: 2/3
Entrer la fraction 2: 3/8
2/3 * 3/8 est 1/4
```



````python
%%writefile main.cpp

#include <iostream>
#include <cmath>
#include <numeric> // pour std::gcd

class Fraction {
	private:
		int m_numerateur{0}, m_denominateur{1};

	public:
    Fraction(int n{0}, int d{1})
        : m_numerateur(n), m_denominateur(d)
    {
        if (m_denominateur < 0) {
            m_numerateur = -m_numerateur;
            m_denominateur = -m_denominateur;
        }

    }

    void print() {
        std::cout << m_numerateur << "/" << m_denominateur << '\n';
    }

    friend Fraction operator*(const Fraction& f1, const Fraction& f2);
    friend Fraction operator*(const Fraction& f, int n);
    friend Fraction operator*(int n, const Fraction& f);
		friend std::ostream& operator<<(std::ostream& os, const Fraction& f);
		friend std::istream& operator>>(std::istream& is, Fraction& f);
};

Fraction operator*(const Fraction& f1, const Fraction& f2) {
    Fraction results(f1.m_numerateur * f2.m_numerateur, f1.m_denominateur * f2.m_denominateur);
    return results;
}

Fraction operator*(const Fraction& f, int n) {
    Fraction results(f.m_numerateur * n, f.m_denominateur);
    return results;
}

Fraction operator*(int n, const Fraction& f) {
    return f * n;
}


int main()
{
	Fraction f1{2, 5};
	f1.print();

	Fraction f2{3, 8};
	f2.print();

	Fraction f3{ f1 * f2 };
	f3.print();

	Fraction f4{ f1 * 2 };
	f4.print();

	Fraction f5{ 2 * f2 };
	f5.print();

	Fraction f6{ Fraction{1, 2} * Fraction{2, 3} * Fraction{3, 4} };
	f6.print();

	Fraction f7{0, 6};
	f7.print();

	return 0;
}
````
