# [C++] 09 - Fonctions - Exercice 1

Ecrire une fonction find() permettant de trouver une valeur dans un tableau trié de type C. On implémentera l'algorithme de recherche binaire (binary search) :


*   Considérer l'élément médian du tableau
*   Si l'élément médian est plus grand que la valeur cible, ne plus prendre en compte la moitié supérieure du tableau
*   Si l'élément médian est plus petit que la valeur cible, ne plus prendre en compte la moitié inférieure du tableau
*   Si l'élément médian est égal à la valeur cible, retourner l'indice de cet élément
*   Si l'élément médian n'a pas été trouvé, retourner le code d'erreur -1

Le programme principal teste la fonction find() pour différentes valeurs à trouver dans un tableau donné.




#Correction 20/20


```python
%%writefile main.cpp

#include <iostream>
#include <iterator>

// tab est le tableau dans lequel la recherche est faite
// valeurCible est la valeur à rechercher (existe ou non)
// indiceMin est l'indice minimum de l'intervalle de recherche
// indiceMax est l'indice maximum de l'intervalle de recherche
// find() retourne l'indice de la valeur cible si elle est trouvée, -1 sinon
int find(const int* tab, int valeurCible, int indiceMin, int indiceMax)
{
    do {
      
      int indiceMedian{(indiceMin + indiceMax) / 2}; // Valeur de l'indice du milieu

      if (tab[indiceMedian] == valeurCible)
          return indiceMedian;
      else if (tab[indiceMedian] < valeurCible)
          indiceMin = indiceMedian + 1;
      else
        indiceMax = indiceMedian - 1;
    } while (indiceMin <= indiceMax);

    return -1;

}

int main()
{
    constexpr int tab[]{ 3, 6, 8, 12, 14, 17, 20, 21, 26, 32, 36, 37, 42, 44, 48 };

    // Test du résultat sur un ensemble de valeurs
    constexpr int numValeursTest{ 9 };
    constexpr int valeursTest[numValeursTest]{ 0, 3, 12, 13, 22, 26, 43, 44, 49 };
    int resultatAttendu[numValeursTest]{ -1, 0, 3, -1, -1, 8, -1, 13, -1 };

    // Boucle sur les valeurs de test
    for (int i{ 0 }; i < numValeursTest; ++i)
    {
        int index{ find(tab, valeursTest[i], 0, static_cast<int>(std::size(tab)) - 1) };
        if (index == resultatAttendu[i])
             std::cout << "valeur test " << valeursTest[i] << " OK\n";
        else
             std::cout << "valeur test " << valeursTest[i] << " NOK\n";
    }

    return 0;
}
```
