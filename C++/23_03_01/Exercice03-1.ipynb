# [C++] 03 - Types de base - Exercice 1
Ecrire un programme qui effectue des opérations entre deux valeurs flottantes.

Le programme demande à l'utilisateur de saisir deux valeurs entières, puis l'un des symboles +, -, * ou / qui détermine le type d'opération à réaliser. 

Le résultat est affiché sous la forme : x + y = z.

On veillera à encapsuler la saisie et l'affichage dans des fonctions comme dans l'exercice [C++] 02 - 1.

```python
%%writefile main.cpp

#include <iostream>

// Déclaration de la fonction 
float operation(float a, float b, char o)
{ 
  float r ;

  switch(o)
  { 
    case '-' : r = a - b ; break;

    case '+' : r = a + b ; break;

    case '*' : r = a * b ; break ;

    case '/' : r = a / b ; break ;

    default : r = a + b ;
  }

  return r;
}

float print(float a, float b, float r)
{
  std::cout << "The result of the choosing operation is : \n";
  std::cout << a <<" + " << b <<" = " << r;

  return r;
}


int main(void)
{
  float nber_1 = {0}, nber_2 = {0}, result = {0};
  char operation_char;

  std::cout << "====== ¨PROGRAMM TO MAKE AN OPERATION WITH TWO DECIMAL VALUES ====== \n";
  
  std::cout << "Enter the first decimal value : \n";
  std::cin >> nber_1;
  std::cout << "Enter the second decimal value : \n";
  std::cin >> nber_2;

  std::cout << "(-), (+), (*), (/) \n";
  std::cout << "Enter one of these symbols to choose the operation to realize : \n";
  std::cin >> operation_char;
  
  result = operation(nber_1, nber_2, operation_char);
  
  print(nber_1, nber_2, result);

  return 0;
}
```
