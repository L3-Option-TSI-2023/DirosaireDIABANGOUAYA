Ecrire un programme qui comporte une fonction déterminant si une année est bissextile.

Une année bissextile est divisible par 4 mais non par 100, ou divisible par 400.
La fonction dispose en entrée d’un paramètre : a ; et retourne : la valeur 1 si l’année est bissextile ; la valeur 0 sinon.

Tester cette fonction avec une fonction main qui :
- permet la saisie d’une année (variable annee)
- appelle la fonction bissextile
- édite le message : année bissextile ou année non bissextile

**Années test :** 2004 (1) 2003 (0) 2000 (1) 1900 (0) 1600(1) 1800 (0)

#Correction 14/20

Ne fonctionne pas pour 1800 et 1900 car une année bissextile n'est pas divisible par 100.

```python
%%writefile main.cpp

#include <iostream>

int bissextile_year(int a)
{
  
  if (a%4==0 || a%400==0) { return 1; }
  else { return 0; }

  return 0;
}

int main(void)
{
  int year={0};

  std::cout<<"2004("<<bissextile_year(2004)<<") ";
  std::cout<<"2003("<<bissextile_year(2003)<<") ";
  std::cout<<"2000("<<bissextile_year(2000)<<") ";
  std::cout<<"1900("<<bissextile_year(1900)<<") ";
  std::cout<<"1600("<<bissextile_year(1600)<<") ";
  std::cout<<"1800("<<bissextile_year(1800)<<") ";

  return 0;
}
```
