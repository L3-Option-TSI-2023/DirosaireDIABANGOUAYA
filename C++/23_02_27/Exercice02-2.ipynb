# [C++] 02 - Fonctions et fichiers - Exercice 2
Reprendre l'exercice [C++] 02 - 1 en déplaçant les fonctions de saisie et d'affichage dans un fichier es.cpp

```python
%%writefile es.cpp


// Saisie des nombres sans precision de type
float writte(std::string directive="")
{
  float unkown_type_nber;

  std::cout << directive << "\n";
  std::cin >> unkown_type_nber;
  return unkown_type_nber;
}


// Affichage de la somme
float print(int a, float b) 
{ 
  float sum{ static_cast<float>(a+b) };

  std::cout << "\nVous avez entré les valeurs suivantes : "
            << a << " et " << b
            << "\nLeur somme fait : " << sum;
  return 0;
}
```

```python

%%writefile main.cpp

#include <iostream>


int main(void)
{
  int nber_1{ 
    static_cast<int>(writte("Saisissez un chiffre/nombre entier :")) 
  }; 

  float nber_2{ 
    static_cast<float>(writte("Saisissez un chiffre/nombre decimal :")) 
  };

  print(nber_1, nber_2);

  return 0;
}
```
