# [C++] 06 - Flux d'exécution et gestion des erreurs - Exercice 2
Compléter le programme ci-dessous en écrivant le corps de la fonction estPremier qui teste si un nombre est premier.

Si le programme fonctionne, les assertions de la fonction main doivent être vérifiées.

```python
%%writefile main.cpp
#include <iostream>
#include <cassert>

// fonction qui teste si le nombre est premier 
bool estPremier(int x)
{
  // Je commence par vérifier que le nombre n'est pas <1 car ca voudrait dire qu'il n'est pas premier
  // d'où je retorune le false

  if (x <= 1)  
    return false;

  // la condition de la divisibilité par 1 étant déjà tester, il me faut maintenant de rassuer que de 2 à l'entier à vérifier,
  // il n'existe aucun diviseur qui génère un reste si c'est le cas l'entier n'est pas premier  
  for (int i = {2}; i < x; ++i)
  {
    if (x % i == 0)
        return false;
  }

  // sinon le nombre est premier     
  return true;
}

int main()
{
    assert(!estPremier(0));
    assert(!estPremier(1));
    assert(estPremier(2));
    assert(estPremier(3));
    assert(!estPremier(4));
    assert(estPremier(5));
    assert(estPremier(7));
    assert(!estPremier(9));
    assert(estPremier(11));
    assert(estPremier(13));
    assert(!estPremier(15));
    assert(!estPremier(16));
    assert(estPremier(17));
    assert(estPremier(19));
    assert(estPremier(97));
    assert(!estPremier(99));
    assert(estPremier(13417));

    std::cout << "Success!\n";

    return 0;
}
```
