# [C++] 05 - Portée, durée et lien - Exercice 1
Ecrire un programme qui demande une valeur à l'utilisateur et qui la teste par rapport à une valeur maximale autorisée.

La valeur maximale sera stockée comme une constante globale au programme dans un fichier d'en-tête constants.h.

```python
%%writefile constants.h

// les instructions "ifndef" "define" & "endif"
// sont utilisées pour empêcher le compilateur d'inclure plusieurs fois ce fichier 

#ifndef CONSTANTS_H
#define CONSTANTS_H

// je déclare cette constante avec "constexpr" car la valeur de NMAX doit être
// connue à la conpilation  

constexpr int NMAX = {10};

#endif
```

```python
%%writefile main.cpp

#include "constants.h"
#include <iostream>


//déclaration de la fonction la valeur par rapport à la constante globaeglobale 
int test_integer(int x)
{
  if (x<NMAX) 
    std::cout<< x <<" < " << NMAX <<"\n";
    else
      std::cout<< x <<" > " << NMAX <<"\n";

  return x;
}

int main(void)
{
    int value;

    std::cout<< "Entrer un entier : ";
    std::cin>> value;

    std::cout<< test_integer(value);

    return 0;
}
```

```python
%%shell
g++ main.cpp -o output
./output
```
