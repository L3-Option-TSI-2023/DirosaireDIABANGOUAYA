Ecrire un programme qui appelle plusieurs fois une même fonction. Cette fonction devra retourner le booléen vrai lors des trois premiers appels, puis le booléen faux lors des appels suivants.

```python
%%writefile main.cpp

#include <iostream>

// déclaration de la foction qui compte le nombre d'auto appel
bool self_call ()
{
  static int counter = {0};
  std::cout<<"\n" << "iteration "<< counter << " : ";

  counter++;

  if(counter<=3)
    return true;
      else
        return false;
}

int main(void)
{
  int i;
 
  for(i = 0; i < 5; i++)
  {
    std::cout << self_call();
  }

  return 0;
}
```

```python
%%shell
g++ main.cpp -o output
./output
```
