Ecrire une fonction template qui prend un nombre $x$ de type quelconque en et un entier $n$ en entrée et qui retourne la somme des $n$ premiers multiples de $x$ : $1*x+2*x+...+n*x$.

L'appeler dans le programme principal pour plusieurs types de données.






```python
%%writefile main.cpp
#include <iostream>

// je declare un template puis la fonction qui avec 
// le paramètre x est de type T ce qui  permettra au complilateur de determiner 
// automatiquement son type une fois sa valeur renseigée

template <typename T>
T duplicate_by(T x, int n)
{
  int sum ={0}, i={0};

  for (i={1}, sum=0 ; i<=n ; i++) 
    sum += x*i ; 
    // l'opération réalisée est la suivante sum=sum+(x*i) : on garde en mémoire 
    // la valeur de l'itération précédente 
    // à laquelle on ajoute la valeur suivante

  return sum;
}

int main(void)
{
  int i={0}, n={0}, value={0};

  std::cout<< "Entrer un nbre : ";
    std::cin>> value;
  std::cout<< "Entrer une valeur max : ";
    std::cin>> n;


  std::cout << "Somme des produits des " 
            << n 
            << " premiers termes par " 
            << value << " est : "<< duplicate_by(value, n) ;

  return 0;
}
```
