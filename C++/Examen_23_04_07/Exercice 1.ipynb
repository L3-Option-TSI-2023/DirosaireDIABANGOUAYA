Ecrire une fonction template `mediane()` qui calcule la valeur médiane d'un tableau dynamique `std::vector` avec un type d'éléments quelconque.

Le programme principal suivant devrait fonctionner.


#Correction 7/7

```python
%%writefile mediane.cpp

#include <vector>
#include <algorithm>

using namespace std;

// Déclaration et initialisation de la fonction template
template<typename T>
T mediane(std::vector<T> vectorTab)
{
  sort(vectorTab.begin(), vectorTab.end()); // Classement par ordre croissant

  if (vectorTab.size() % 2 == 0)
    return (vectorTab[vectorTab.size() / 2 - 1] + vectorTab[vectorTab.size() / 2]) / 2;
  else
    return vectorTab[vectorTab.size() / 2];
}
```

```python
%%writefile main.cpp

#include <iostream>
#include <vector>
#include <algorithm>
#include "mediane.cpp"


int main()
{
  std::vector<double> tabDouble{ 5.2, 3.4, 9.1, 1.5, 2.7, 2.2, 4.4 };
  std::cout << "La valeur médiane du tableau en virgule flottante est " << mediane<>(tabDouble) << '\n';

  std::vector<double> tabInt{ 5, 3, 9, 1, 2, 2, 4 };
  std::cout << "La valeur médiane du tableau en entier est " << mediane<>(tabInt) << '\n';

  return 0;
}
```
