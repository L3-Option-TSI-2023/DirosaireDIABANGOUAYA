# [C++] Examen - Exercice 2

Ecrire une classe abstraite `Forme` qui comporte une fonction virtuelle pure `aire()` permettant de calculer l'aire de la surface de la forme. `Triangle` est une classe qui dérive de la classe `Forme` et qui comporte deux variables membre pour stocker la base et la hauteur.

Le programme principal suivant devrait fonctionner.

#Correction 7/7

```python
%%writefile main.cpp

#include <iostream>
#include <cmath>


class Forme {

  public:
    virtual double aire() const = 0; // Fonction virtuelle pour calculer l'aire
    virtual ~Forme() {} // Destructeur virtuel pour détruire les objets
};

class Triangle : public Forme {

  public:
    Triangle(double b, double h) : m_base(b), m_hauteur(h) {} // Constructeur de la classe fille Triangle initialisation de ligne

    // const car elles ne modifient pas les attributs de la classe
    double aire() const 
    { 
      return (m_base * m_hauteur) / 2; // Calcul de l'aire du triangle
    } 

    double getBase() const 
    { 
      return m_base; 
    }

    double getHauteur() const 
    { 
      return m_hauteur; 
    }
    
  private:
    double m_base;
    double m_hauteur;
};


int main(void)
{
  Triangle triangle( 6.2, 3.4 );
  Forme& forme{ triangle };
  std::cout << "Le triangle de base " << triangle.getBase() << " et de hauteur " << triangle.getHauteur() << " a pour aire " << forme.aire() << '\n';

  return 0;
}
```
