# [C++] 08 - Tableaux, chaines, pointeurs et références - Exercice 1
Ecrire une fonction permettant de retourner un tableau statique rempli avec des valeurs aléatoires (utiliser std::rand()), puis une autre fonction retournant le maximum des valeurs que contient le tableau. 

On considèrera des fonctions template et non template (type int), et des tableaux statiques de type C et std::array.

Dans le détail, les fonctions à écrire sont :

1.   randomArrayCInt & maxArrayCInt : tableau statique de type C (int)
2.   randomArrayC & maxArrayC : tableau statique de type C (template)
3.   randomArrayInt & maxArrayInt : tableau statique std::array (int)
4.   randomArray & maxArray : tableau statique std::array (template)







```python
%%writefile main.cpp
#include <iostream>
#include <array>

constexpr int N{10};

// **************************
int* randomArrayCInt()
{
  //static pour que apres le return il continue a exister
  static int tab[N];

  //parcourir le tableau pour add random values
  for (int i{0}; i<N; i++)
    tab[i] = std::rand() % 100;
  return tab;
}
// **************************

// **************************
int maxArrayCInt(int* tab) 
{
  // on initialise a partir de l'indice  pour parcourir à partir d'elle
  int maxInt {tab[0]};

  for (int i{1}; i<N; ++i)
    if (tab[i] > maxInt)
      maxInt = tab[i];

  return maxInt;
}
// **************************

// **************************
template <typename T>
T* randomArrayC()
{
  static T tab[N];

  for (int i{}; i<N; ++i)
    tab[i] = std::rand() % 100;

  return tab;
}
// **************************

// **************************
template <typename T>
T maxArrayC(T* tab)
{
  T max{ tab[0]};

  for(int i{1}; i<N; ++i)
    if (tab[i] > max)
      max = tab[i];
  return max;
}
// **************************


// 3 **************************
std::array<int, N> randomArrayInt()
{
  std::array<int, N> tab;

  for (int i{}; i <N; ++i)
    tab[i] = std::rand() % 100;

    return tab;
}
// **************************

// 3 **************************
int maxArrayInt(const std::array<int, N>& tab)
{
  int maxInt{ tab[0] };

  for (int i{1}; i<N; ++i)
    maxInt = tab[i];
  
  return maxInt;
}
// **************************

// 4 **************************
template<typename T>
std::array<T, N> randomArray() 
{
  std::array<T, N> tab;

  for (int i{}; i<N; ++i)
    tab[i] = std::rand() % 100;
  
  return tab;
}
// **************************

// 4 **************************
template <typename T>
T maxArray(const std::array<T, N>& tab)
{
  T max{ tab[0] };
  for (int i{1}; i<N; ++i)
    max = tab[i];
  return max;
}
// **************************

int main(void)
{
  // 1 tab int de type C
  auto tabCInt {randomArrayCInt()};

  for (int i{}; i<N; ++i)
    std::cout << tabCInt[i] << ' ';
  std::cout << '\n';
  std::cout << "Le maximum du tableau est : " << maxArrayCInt(tabCInt) << '\n';

  // 2 - tableau template de type C
  auto tabC{ randomArrayC<double>() };
  std::cout <<"Le maximum du tableau est : " << maxArrayC(tabC) << '\n';

  // 3 - tableau int de type std::array
  auto tabInt { randomArrayInt() };
  std::cout <<"Le maximum du tableau est : " << maxArrayInt(tabInt) << '\n';

  // 4 - tableau template de type std::array
  auto tab { randomArray<double>() };
  std::cout <<"Le maximum du tableau est : " << maxArray(tab) << '\n';

  return 0;
}
```
